## Model

```python
import xgboost as xgb

from hyperopt import hp, tpe
from hyperopt.fmin import fmin

from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import make_scorer
```

```python
def generate_datesets(
    df, val_pdb_num=500, val_sample_size=5000, train_sample_size=30000
):

    cdf = df[
        df[df.columns[5:-1]]
        .apply(lambda x: x.between(x.min(), x.quantile(0.995)))
        .all(1)
    ]

    pdb_ids_cut = list(np.random.choice(cdf["PDBID"].unique(), size=val_pdb_num))
    val_cut = cdf.query(f"PDBID in {pdb_ids_cut}")

    bins = [i / 20 for i in range(20 + 1)]
    group = val_cut.groupby(pd.cut(val_cut.Fitness, bins))
    val_set = group.apply(
        lambda df: df.sample(val_sample_size, replace=True) if len(df) > 0 else df
    )

    val_set_pdb_id = list(val_set.PDBID.unique())
    val_set_seq_id = list(val_set.SEQID.unique())

#     train_set = cdf.query(f"PDBID not in {val_set_pdb_id}")
    train_set = cdf.query(
        f"PDBID not in {val_set_pdb_id} and SEQID not in {val_set_seq_id}"
    )

    train_set = train_set.groupby(pd.cut(train_set.Fitness, bins)).apply(
        lambda df: df.sample(train_sample_size, replace=True) if len(df) > 0 else df
    )

    x_train, y_train = train_set.loc[:, train_set.columns[2:-1]], train_set["Fitness"]
    x_val, y_val = val_set.loc[:, val_set.columns[2:-1]], val_set["Fitness"]

    #     mean, std = x_train.mean(), x_train.std()
    #     x_train = (x_train - mean) / std
    #     x_val = (x_val - mean) / std

    return (x_train, y_train), (x_val, y_val)
```

```python
(x_train, y_train), (x_val, y_val) = generate_datesets(df, 400, 50, 30000)
```

```python
x_train.shape, x_val.shape
```

```python
mae = lambda yt, yp: np.mean(np.abs(yt - yp))
mae_scorer = make_scorer(mae, greater_is_better=False)
```

```python
def objective_lr(params):
    params = {
        "learning_rate": float(params["learning_rate"]),
        "n_estimators": int(params["n_estimators"]),
    }

    clf = xgb.XGBRegressor(
        objective="binary:logistic", tree_method="gpu_hist", gpu_id=0, **params
    )

    score = -cross_val_score(
        clf, x_train, y_train, scoring="neg_mean_absolute_error", cv=5
    ).mean()
    print("MAE {:.3f} params {}".format(score, params))
    return score


space = {
    "learning_rate": hp.uniform("learning_rate", 0.5, 0.01),
    "n_estimators": hp.quniform("n_estimators", 100, 2000, 10),
}

best_lr = fmin(fn=objective_lr, space=space, algo=tpe.suggest, max_evals=20)
print("Hyperopt estimated optimum {}".format(best_lr))
```

```python
def objective_other(params):
    params = {
        "max_depth": int(params["max_depth"]),
        "gamma": "{:.3f}".format(params["gamma"]),
        "colsample_bytree": "{:.3f}".format(params["colsample_bytree"]),
        "min_child_weight": "{:.2f}".format(params["min_child_weight"]),
    }

    clf = xgb.XGBRegressor(
        objective="binary:logistic",
        n_estimators=1500,
        learning_rate=0.045,
        tree_method="gpu_hist",
        gpu_id=0,
        **params
    )

    score = -cross_val_score(
        clf, x_train, y_train, scoring="neg_mean_absolute_error", cv=5
    ).mean()
    print("MAE {:.3f} params {}".format(score, params))
    return score


space = {
    "max_depth": hp.quniform("max_depth", 1, 10, 1),
    "colsample_bytree": hp.uniform("colsample_bytree", 0.3, 1.0),
    "gamma": hp.uniform("gamma", 0.0, 0.5),
    'min_child_weight': hp.quniform('min_child_weight', 1, 6, 1),
}

best = fmin(fn=objective_other, space=space, algo=tpe.suggest, max_evals=20)
```

```python
print("Hyperopt estimated optimum {}".format(best))
```

```python
random_state = 42


def objective(params):
    params["n_estimators"] = int(params["n_estimators"])
    params["max_depth"] = int(params["max_depth"])

    reg = xgb.XGBRegressor(
        objective="binary:logistic",
        tree_method="gpu_hist",
        gpu_id=0,
        nthread=4,
        seed=random_state,
        early_stopping_rounds=20,
        **params
    )

    score = -cross_val_score(
        reg, x_train, y_train, scoring="neg_mean_absolute_error", cv=5
    ).mean()
    print("MAE {:.3f} params {}".format(score, params))
    with open('log.txt', 'a') as f:
         print("MAE {:.3f} params {}".format(score, params), file=f)
    return score


space = {
    "learning_rate": hp.uniform("learning_rate", 0.01, 0.5),
    "n_estimators": hp.quniform("n_estimators", 200, 2000, 10),
#     "eta": hp.quniform("eta", 0.025, 0.5, 0.01),
    # A problem with max_depth casted to float instead of int with
    # the hp.quniform method.
    "max_depth": hp.quniform("max_depth", 1, 15, 1),
    "min_child_weight": hp.quniform("min_child_weight", 1, 6, 1),
    "subsample": hp.quniform("subsample", 0.5, 1, 0.05),
    "gamma": hp.quniform("gamma", 0, 1, 0.01),
    "colsample_bytree": hp.quniform("colsample_bytree", 0.1, 1, 0.05),
}

best = fmin(fn=objective, space=space, algo=tpe.suggest, max_evals=200)
```

```python
def plot_tree(xgb_model, num_trees, filename, rankdir="UT"):
    """
    Plot the tree in high resolution
    :param xgb_model: xgboost trained model
    :param filename: the pdf file where this is saved
    :param rankdir: direction of the tree: default Top-Down (UT), accepts:'LR' for left-to-right tree
    :return:
    """
    import xgboost as xgb
    import os

    if num_trees is None:
        num_trees = xgb_model.best_iteration
    gvz = xgb.to_graphviz(xgb_model, num_trees=num_trees, rankdir=rankdir)
    _, file_extension = os.path.splitext(filename)
    format = file_extension.strip(".").lower()
    data = gvz.pipe(format=format)
    full_filename = filename
    with open(full_filename, "wb") as f:
        f.write(data)
```
