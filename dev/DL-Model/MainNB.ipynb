## Load

```python
import pandas as pd
import numpy as np

DATASET_PATH = "dataset.csv"
```

```python
AAlist = [
    "A",
    "R",
    "N",
    "D",
    "C",
    "Q",
    "E",
    "G",
    "H",
    "I",
    "L",
    "K",
    "M",
    "F",
    "P",
    "S",
    "T",
    "W",
    "Y",
    "V",
]

columns = (
    ["PDBID", "SEQID", "Length", "Solvent", "SecondaryS", "Alignment", "DFIRE2"]
    + ["Pot_" + AA for AA in AAlist]
    + ["Freq_" + AA for AA in AAlist]
    + ["Fitness"]
)
```

```python
df = pd.read_csv(DATASET_PATH, names=columns)
```

```python
df.sample(20)
```

```python
df.Fitness.plot(kind="hist")
```

## Train/Test split

```python
df["PDBID"].unique().shape
```

```python
df["SEQID"].unique().shape
```

```python
pdb_ids_cut = list(np.random.choice(df["PDBID"].unique(), size=1000))
```

```python
val = df.query(f"PDBID in {pdb_ids_cut}")
```

```python
bins = [i / 10 for i in range(11)]
```

```python
group = val.groupby(["Length", pd.cut(val.Fitness, bins)])
```
